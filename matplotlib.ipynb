```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=[171,151,124,134,156]
y=[80,60,45,50,65]
mean_x=np.mean(x)
mean_y=np.min(y)
nuem=0
denom=0
n=len(x)
for i in range (n):
    nuem+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
m=nuem/denom
c=mean_y-(m*mean_x)
# plt.scatter(x,y,color="red")
# plt.plot(x,y,color="blue")
max_x=np.max(x) + 100
min_x=np.min(x) - 100
a=np.linspace(min_x,max_x,100)
b=m*a+c
print(b)
plt.scatter (x,y,color="red")
plt.plot(a,b,color="green")
plt.show()
```
